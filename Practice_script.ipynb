```python
# Join the columns
AgeDF['AgeClass'] = AgeDF["ID"] + AgeDF["age"]
```

```python
# Drop unwanted columns
AgeDF = AgeDF.drop(columns=['ID', 'age'])
```

```python
# Create another column with size of AGEID groups
s = pd.Series(['0','5'])
AgeDF.insert(loc=1, column = 'Size',value=s)
AgeDF.head()
```
